# Tutorial de sympy para multiplicar matrices

```python
import sympy as sp
```

Como representar matrices:

```python
A = sp.Matrix([[1, 2], [3, 4]])
display(A)
```

Se puede crear matrices más grandes o con variables simbólicas de la misma manera:

```python
x, y, z = sp.symbols('x y z')
B = sp.Matrix([[x, y], [z, 0]])
display(B)
```

Obtener una entrada específica

```python
# Para obtener un elemento específico, usa la indexación [fila, columna] (base 0).
# Por ejemplo, para obtener el elemento en la primera fila, segunda columna (valor 2):
elemento_1_2 = A[0, 1]
display(f"Elemento en (0, 1): {elemento_1_2}")

# Para obtener el elemento en la segunda fila, primera columna (valor 3):
elemento_2_1 = A[1, 0]
display(f"Elemento en (1, 0): {elemento_2_1}")
```

Obtener una fila específica

```python
# Para obtener una fila completa, usa el método .row() o slice.
# Por ejemplo, para obtener la primera fila (índice 0):
primera_fila = A.row(0)
display(f"Primera fila: {primera_fila}")

# O usando slicing (devuelve un objeto Matrix con una fila):
primera_fila_slice = A[0, :]
display(f"Primera fila (slice): {primera_fila_slice}")
```

Obtener una columna específica

```python
# Para obtener una columna completa, usa el método .col() o slice.
# Por ejemplo, para obtener la primera columna (índice 0):
primera_columna = A.col(0)
display(f"Primera columna: {primera_columna}")

# O usando slicing (devuelve un objeto Matrix con una columna):
primera_columna_slice = A[:, 0]
display(f"Primera columna (slice): {primera_columna_slice}")
```

Multiplicación de matrices

```python
# Definimos dos matrices para la multiplicación
M1 = sp.Matrix([[1, 2], [3, 4]])
M2 = sp.Matrix([[5, 6], [7, 8]])

display("Matriz M1:", M1)
display("Matriz M2:", M2)

# Realizamos la multiplicación de matrices
resultado = M1 * M2

display("Resultado de M1 * M2:", resultado)
```

Asegúrate de que el número de columnas de la primera matriz coincida con el número de filas de la segunda matriz para que la multiplicación sea posible.

Elevar una matriz a una potencia

```python
# Para elevar una matriz a una potencia, usa el operador **
# Por ejemplo, para A al cuadrado (A^2):
potencia_2 = A ** 2
display(f"Matriz A:\n", A)
display(f"A elevada a la potencia 2 (A**2):\n", potencia_2)

# También puedes usar el método .pow()
potencia_3 = A.pow(3)
display(f"A elevada a la potencia 3 (A.pow(3)):\n", potencia_3)
```

Tener en cuenta que la matriz debe ser cuadrada para poder elevarla a una potencia.

Multiplicación de matriz por vector

```python
# Definimos una matriz y un vector para la multiplicación
matriz_m = sp.Matrix([[1, 2, 3], [4, 5, 6]])
vector_v = sp.Matrix([[7], [8], [9]])

display("Matriz M:", matriz_m)
display("Vector V:", vector_v)

# Realizamos la multiplicación de la matriz por el vector
resultado_mv = matriz_m * vector_v

display("Resultado de M * V:", resultado_mv)
```

Para que la multiplicación sea posible, el número de columnas de la matriz (`matriz_m` tiene 3 columnas) debe ser igual al número de filas del vector (`vector_v` tiene 3 filas). El resultado será un nuevo vector con el número de filas de la matriz (`matriz_m` tiene 2 filas) y 1 columna.

### Hallar Valores y Vectores Propios

```python
# Definimos una matriz cuadrada para encontrar sus valores y vectores propios
matriz_cuadrada = sp.Matrix([[2, 1], [1, 2]])
display("Matriz para eigenvalores/eigenvectores:", matriz_cuadrada)

# Calcular los valores propios (eigenvalues)
eigenvalores = matriz_cuadrada.eigenvals()
display("Valores propios:", eigenvalores)

# Calcular los vectores propios (eigenvectors)
eigenvectores = matriz_cuadrada.eigenvects()
display("Vectores propios:", eigenvectores)
```

El método `eigenvals()` devuelve un diccionario donde las claves son los valores propios y los valores son sus multiplicidades algebraicas.

El método `eigenvects()` devuelve una lista de tuplas en el formato `(valor_propio, multiplicidad_algebraica, [lista_de_vectores_propios_correspondientes])`.

# Dibujar un Diagrama de Transición con NetworkX

Primero, necesitamos importar las librerías necesarias: `networkx` para la creación y manipulación de grafos, y `matplotlib.pyplot` para la visualización.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Luego, definimos los estados y las transiciones de nuestro sistema. Para un diagrama de transición, usaremos un grafo dirigido (`DiGraph`), ya que las transiciones van de un estado a otro en una dirección específica.

```python
# Crear un grafo dirigido
G = nx.DiGraph()

# Definir los estados (nodos)
estados = ['Inicio', 'Estado A', 'Estado B', 'Fin']
G.add_nodes_from(estados)

# Definir las transiciones (aristas)
# Cada tupla es (estado_origen, estado_destino, {atributos})
transiciones = [
    ('Inicio', 'Estado A', {'label': 'Evento 1'}),
    ('Estado A', 'Estado B', {'label': 'Evento 2'}),
    ('Estado B', 'Estado A', {'label': 'Evento 3'}),
    ('Estado A', 'Fin', {'label': 'Evento 4'}),
    ('Estado B', 'Fin', {'label': 'Evento 5'})
]

G.add_edges_from(transiciones)
```

Finalmente, dibujamos el grafo. Podemos personalizar la apariencia de los nodos, las aristas y las etiquetas para que el diagrama sea claro y representativo.

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# Definir la posición de los nodos (layout)
# spring_layout es una buena opción para visualizaciones generales
pos = nx.spring_layout(G, seed=42) # seed para reproducibilidad

# Dibujar los nodos
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color='lightblue', alpha=0.9)

# Dibujar las aristas
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray', width=2)

# Dibujar las etiquetas de los nodos
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# Dibujar las etiquetas de las aristas (eventos)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=9)

# Ocultar los ejes de matplotlib
plt.axis('off')

# Añadir un título
plt.title('Diagrama de Transición de Ejemplo', size=15)

# Mostrar el diagrama
plt.show()
```

**Explicación del código:**

1.  **`import networkx as nx` y `import matplotlib.pyplot as plt`**: Importa las librerías necesarias.
2.  **`G = nx.DiGraph()`**: Crea un objeto de grafo dirigido. Esto es crucial para los diagramas de transición, ya que las relaciones entre estados son unidireccionales.
3.  **`estados = [...]` y `G.add_nodes_from(estados)`**: Define una lista de estados y los añade como nodos al grafo.
4.  **`transiciones = [...]` y `G.add_edges_from(transiciones)`**: Define las transiciones como tuplas `(origen, destino, {atributos})`. El atributo `'label'` se usa para describir el evento que causa la transición.
5.  **`plt.figure(figsize=(8, 6))`**: Establece el tamaño de la figura para la visualización.
6.  **`pos = nx.spring_layout(G, seed=42)`**: Calcula la posición de cada nodo en el diagrama. `spring_layout` es un algoritmo de diseño de grafos que intenta colocar los nodos de manera que las aristas tengan una longitud uniforme y minimicen los cruces. `seed` se usa para asegurar que la disposición sea la misma cada vez que se ejecute el código.
7.  **`nx.draw_networkx_nodes(...)`**: Dibuja los nodos del grafo, permitiendo personalizar su tamaño, color y transparencia.
8.  **`nx.draw_networkx_edges(...)`**: Dibuja las aristas con flechas para indicar la dirección de la transición, configurando el color y el grosor.
9.  **`nx.draw_networkx_labels(...)`**: Dibuja las etiquetas (nombres) de los nodos.
10. **`nx.get_edge_attributes(G, 'label')` y `nx.draw_networkx_edge_labels(...)`**: Obtiene las etiquetas de las transiciones (eventos) que definimos y las dibuja sobre las aristas.
11. **`plt.axis('off')` y `plt.title(...)`**: Oculta los ejes de Matplotlib y añade un título al diagrama.
12. **`plt.show()`**: Muestra el diagrama generado.

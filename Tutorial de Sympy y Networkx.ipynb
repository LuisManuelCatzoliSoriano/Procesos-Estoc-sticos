# Tutorial de sympy para multiplicar matrices

```python
import sympy as sp
```

Como representar matrices:

```python
A = sp.Matrix([[1, 2], [3, 4]])
display(A)
```

Se puede crear matrices más grandes o con variables simbólicas de la misma manera:

```python
x, y, z = sp.symbols('x y z')
B = sp.Matrix([[x, y], [z, 0]])
display(B)
```

Para obtener una entrada específica

```python
# Para obtener un elemento específico, usa la indexación [fila, columna] (base 0).
# Por ejemplo, para obtener el elemento en la primera fila, segunda columna (valor 2):
elemento_1_2 = A[0, 1]
display(f"Elemento en (0, 1): {elemento_1_2}")

# Para obtener el elemento en la segunda fila, primera columna (valor 3):
elemento_2_1 = A[1, 0]
display(f"Elemento en (1, 0): {elemento_2_1}")
```

Para obtener una fila específica

```python
# Para obtener una fila completa, usa el método .row() o slice.
# Por ejemplo, para obtener la primera fila (índice 0):
primera_fila = A.row(0)
display(f"Primera fila: {primera_fila}")

# O usando slicing (devuelve un objeto Matrix con una fila):
primera_fila_slice = A[0, :]
display(f"Primera fila (slice): {primera_fila_slice}")
```

Para obtener una columna específica

```python
# Para obtener una columna completa, usa el método .col() o slice.
# Por ejemplo, para obtener la primera columna (índice 0):
primera_columna = A.col(0)
display(f"Primera columna: {primera_columna}")

# O usando slicing (devuelve un objeto Matrix con una columna):
primera_columna_slice = A[:, 0]
display(f"Primera columna (slice): {primera_columna_slice}")
```

Multiplicación de matrices

```python
# Definimo dos matrices para la multiplicación
M1 = sp.Matrix([[1, 2], [3, 4]])
M2 = sp.Matrix([[5, 6], [7, 8]])

display("Matriz M1:", M1)
display("Matriz M2:", M2)

# Realizo la multiplicación de matrices
resultado = M1 * M2

display("Resultado de M1 * M2:", resultado)
```

Se debe asegurar que el número de columnas de la primera matriz coincida con el número de filas de la segunda matriz para que la multiplicación se pueda realizar.

Elevar una matriz a una potencia

```python
# Para elevar una matriz a una potencia, usa el operador **
# Por ejemplo, para A al cuadrado (A^2):
potencia_2 = A ** 2
display(f"Matriz A:\n", A)
display(f"A elevada a la potencia 2 (A**2):\n", potencia_2)

# También se puede usar el método .pow()
potencia_3 = A.pow(3)
display(f"A elevada a la potencia 3 (A.pow(3)):\n", potencia_3)
```

La matriz debe ser cuadrada para poder elevarla a una potencia.

Multiplicación de matriz por vector

```python
# Definimo una matriz y un vector para la multiplicación
matriz_m = sp.Matrix([[1, 2, 3], [4, 5, 6]])
vector_v = sp.Matrix([[7], [8], [9]])

display("Matriz M:", matriz_m)
display("Vector V:", vector_v)

# Realizo la multiplicación de la matriz por el vector
resultado_mv = matriz_m * vector_v

display("Resultado de M * V:", resultado_mv)
```

Para que la multiplicación sea posible, el número de columnas de la matriz (`matriz_m` tiene 3 columnas) debe ser igual al número de filas del vector (`vector_v` tiene 3 filas). El resultado será un nuevo vector con el número de filas de la matriz (`matriz_m` tiene 2 filas) y 1 columna.

### Hallar Valores y Vectores Propios

```python
# Definimo una matriz para encontrar sus valores y vectores propios
matriz_c = sp.Matrix([[1, 2], [3, 2]])
display("Matriz para eigenvalores y eigenvectores:", matriz_c)

# Calculo los valores propios
eigenvalores = matriz_c.eigenvals()
display("Valores propios:", eigenvalores)

# Calculo los vectores propios
eigenvectores = matriz_c.eigenvects()
display("Vectores propios:", eigenvectores)
```

El método `eigenvals()` devuelve un diccionario donde las claves son los valores propios y los valores son sus multiplicidades algebraicas.

El método `eigenvects()` devuelve una lista de tuplas en el formato `(valor_propio, multiplicidad_algebraica, [lista_de_vectores_propios_correspondientes])`.

# Dibujar un Diagrama de Transición con NetworkX

Tome un ejemplo de un diagrama que realizamos en clase. Finalmente, el código para crear el diagrama.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Definir los estados (nodos)
estados = ['Estado 1', 'Estado 2', 'Estado 3']
G.add_nodes_from(estados)

# Cada tupla es (estado origen, estado destino, probabilidad)
transiciones = [
    ('Estado 1', 'Estado 1', {'weight': 0.2, 'label': '0.2'}),
    ('Estado 1', 'Estado 2', {'weight': 0.3, 'label': '0.3'}),
    ('Estado 1', 'Estado 3', {'weight': 0.5, 'label': '0.5'}),
    ('Estado 2', 'Estado 1', {'weight': 0.1, 'label': '0.1'}),
    ('Estado 2', 'Estado 2', {'weight': 0.0, 'label': '0'}), # Probabilidad 0
    ('Estado 2', 'Estado 3', {'weight': 0.9, 'label': '0.9'}),
    ('Estado 3', 'Estado 1', {'weight': 0.55, 'label': '0.55'}),
    ('Estado 3', 'Estado 2', {'weight': 0.0, 'label': '0'}), # Probabilidad 0
    ('Estado 3', 'Estado 3', {'weight': 0.45, 'label': '0.45'})
]

# Las transiciones con probabilidad 0 no se deben dibujar
transiciones_filtradas = [t for t in transiciones if t[2]['weight'] > 0]

G.add_edges_from(transiciones_filtradas)

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 8))

# Definir la posición de los nodos
pos = nx.spring_layout(G, seed=42)

# Dibujar los nodos
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color='lightgreen', alpha=0.9)

# Dibujar las aristas
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray', width=2)

# Dibujar las etiquetas de los nodos
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# Dibujar las etiquetas de las aristas (probabilidades) de manera que no se encimen
edge_labels = nx.get_edge_attributes(G, 'label')

for (u, v, d) in G.edges(data=True):
    label = d['label']
    x, y = (pos[u] + pos[v]) / 2 # Posición media de la arista

    # Ajustar la posición
    if u == v:
        x_text, y_text = pos[u]
        if u == 'Estado 1':
            plt.text(x_text + 0.05, y_text + 0.2, label, size=10, color='red', ha='center', va='center', bbox=dict(facecolor='white', alpha=0))
        elif u == 'Estado 3':
             plt.text(x_text - 0.05, y_text + 0.2, label, size=10, color='red', ha='center', va='center', bbox=dict(facecolor='white', alpha=0))
        else: # Se ajusta según sea necesario
            plt.text(x_text, y_text + 0.2, label, size=10, color='red', ha='center', va='center', bbox=dict(facecolor='white', alpha=0))
    else:
        # Para aristas normales
        offset_x, offset_y = 0, 0 # Default sin offset

        if (u,v) == ('Estado 1', 'Estado 2'):
            offset_x, offset_y = -0.1, 0.1 # Mover ligeramente arriba a la izquierda
        elif (u,v) == ('Estado 2', 'Estado 1'):
            offset_x, offset_y = 0.1, -0.1 # Mover ligeramente abajo a la derecha
        elif (u,v) == ('Estado 1', 'Estado 3'):
            offset_x, offset_y = -0.05, 0.05 # Mover ligeramente arriba a la izquierda
        elif (u,v) == ('Estado 3', 'Estado 1'):
            offset_x, offset_y = 0.05, -0.05 # Mover ligeramente abajo a la derecha
        elif (u,v) == ('Estado 2', 'Estado 3'):
            offset_x, offset_y = 0.1, -0.05 # Mover ligeramente abajo a la derecha
        else:
             offset_x, offset_y = 0.05, 0.05

        plt.text(x + offset_x, y + offset_y, label, size=10, color='red', ha='center', va='center', bbox=dict(facecolor='white', alpha=0))

# Ocultar los ejes de matplotlib
plt.axis('off')

# Añadir un título
plt.title('Diagrama de Transición de Ejemplo con NetworkX', size=16)

# Mostrar el diagrama
plt.show()
```
